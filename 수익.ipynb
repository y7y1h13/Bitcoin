```python
import keras
import numpy as np
import pickle

with open( "final.pkl", "rb" ) as file:
    df_change = pickle.load(file)
    
import joblib
file_name = './closeminmaxsacler__final.pkl'
closeminmaxsacler = joblib.load(file_name)
```

```python
def bit(model):
    
    X_train = np.load('./X_train.npy')
    X_valid = np.load('./X_valid.npy')
    y_train = np.load('./y_train.npy')
    y_valid = np.load('./y_valid.npy')
    X_test = np.load('./X_test.npy')
    y_test = np.load('./y_test.npy')

    # 정상 값으로 변경



#     pred = model.predict(X_valid)
    pred = model.predict(X_test)
    
    y_valid = y_test



    asset0_predict = closeminmaxsacler.inverse_transform(pred[:, 10])
    asset1_predict = closeminmaxsacler.inverse_transform(pred[:, 11])
    asset2_predict = closeminmaxsacler.inverse_transform(pred[:, 12])
    asset3_predict = closeminmaxsacler.inverse_transform(pred[:, 13])
    asset4_predict = closeminmaxsacler.inverse_transform(pred[:, 14])
    asset5_predict = closeminmaxsacler.inverse_transform(pred[:, 15])
    asset6_predict = closeminmaxsacler.inverse_transform(pred[:, 16])
    asset7_predict = closeminmaxsacler.inverse_transform(pred[:, 17])
    asset8_predict = closeminmaxsacler.inverse_transform(pred[:, 18])
    asset9_predict = closeminmaxsacler.inverse_transform(pred[:, 19])
    asset10_predict = closeminmaxsacler.inverse_transform(pred[:, 20])
    asset11_predict = closeminmaxsacler.inverse_transform(pred[:, 21])
    asset12_predict = closeminmaxsacler.inverse_transform(pred[:, 22])
    asset13_predict = closeminmaxsacler.inverse_transform(pred[:, 23])

    asset0_real = closeminmaxsacler.inverse_transform(y_valid[:, 10].reshape(-1,1))
    asset1_real = closeminmaxsacler.inverse_transform(y_valid[:, 11].reshape(-1,1))
    asset2_real = closeminmaxsacler.inverse_transform(y_valid[:, 12].reshape(-1,1))
    asset3_real = closeminmaxsacler.inverse_transform(y_valid[:, 13].reshape(-1,1))
    asset4_real = closeminmaxsacler.inverse_transform(y_valid[:, 14].reshape(-1,1))
    asset5_real = closeminmaxsacler.inverse_transform(y_valid[:, 15].reshape(-1,1))
    asset6_real = closeminmaxsacler.inverse_transform(y_valid[:, 16].reshape(-1,1))
    asset7_real = closeminmaxsacler.inverse_transform(y_valid[:, 17].reshape(-1,1))
    asset8_real = closeminmaxsacler.inverse_transform(y_valid[:, 18].reshape(-1,1))
    asset9_real = closeminmaxsacler.inverse_transform(y_valid[:, 19].reshape(-1,1))
    asset10_real = closeminmaxsacler.inverse_transform(y_valid[:, 20].reshape(-1,1))
    asset11_real = closeminmaxsacler.inverse_transform(y_valid[:, 21].reshape(-1,1))
    asset12_real = closeminmaxsacler.inverse_transform(y_valid[:, 22].reshape(-1,1))
    asset13_real = closeminmaxsacler.inverse_transform(y_valid[:, 23].reshape(-1,1))

    asset0_real = asset0_real.astype('float32')
    asset1_real = asset1_real.astype('float32')
    asset2_real = asset2_real.astype('float32')
    asset3_real = asset3_real.astype('float32')
    asset4_real = asset4_real.astype('float32')
    asset5_real = asset5_real.astype('float32')
    asset6_real = asset6_real.astype('float32')
    asset7_real = asset7_real.astype('float32')
    asset8_real = asset8_real.astype('float32')
    asset9_real = asset9_real.astype('float32')
    asset10_real = asset10_real.astype('float32')
    asset11_real = asset11_real.astype('float32')
    asset12_real = asset12_real.astype('float32')
    asset13_real = asset13_real.astype('float32')

    
    
    for i in range(0,10):
        def get_returns(predict, real):    
            deposit = 0
            # compare_predict_real
            # boolean
            buy_asset = real[:-20][predict[10:] > predict[9-i:-1-i]]
            # buy and sell process
            sell_asset = real[20:][predict[10:] > predict[9-i:-1-i]]
            deposit -= np.sum(buy_asset)
            deposit += np.sum(sell_asset)

            # get profit
            returns = deposit
            return returns


        returns_0 = get_returns(asset0_predict, asset0_real)
        returns_1 = get_returns(asset1_predict, asset1_real)
        returns_2= get_returns(asset2_predict, asset2_real)
        returns_3 = get_returns(asset3_predict, asset3_real)
        returns_4 = get_returns(asset4_predict, asset4_real)
        returns_5 = get_returns(asset5_predict, asset5_real)
        returns_6 = get_returns(asset6_predict, asset6_real)
        returns_7 = get_returns(asset7_predict, asset7_real)
        returns_8 = get_returns(asset8_predict, asset8_real)
        returns_9 = get_returns(asset9_predict, asset9_real)
        returns_10 = get_returns(asset10_predict, asset10_real)
        returns_11 = get_returns(asset11_predict, asset11_real)
        returns_12 = get_returns(asset12_predict, asset12_real)
        returns_13 = get_returns(asset13_predict, asset13_real)

        depo = df_change.groupby('Asset_ID')['Close'].max()

        print(returns_0 / (depo[0]*10)*100)
        print(returns_1 / (depo[1]*10)*100)
        print(returns_2 / (depo[2]*10)*100)
        print(returns_3 / (depo[3]*10)*100)
        print(returns_4 / (depo[4]*10)*100)
        print(returns_5 / (depo[5]*10)*100)
        print(returns_6 / (depo[6]*10)*100)
        print(returns_7 / (depo[7]*10)*100)
        print(returns_8 / (depo[8]*10)*100)
        print(returns_9 / (depo[9]*10)*100)
        print(returns_10 / (depo[10]*10)*100)
        print(returns_11 / (depo[11]*10)*100)
        print(returns_12 / (depo[12]*10)*100)
        print(returns_13 / (depo[13]*10)*100)
        print('-------------------------------------------------------------')
```
